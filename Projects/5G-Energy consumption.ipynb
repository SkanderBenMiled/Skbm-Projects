```python
df['Time'] = pd.to_datetime(df['Time'], format='%Y%m%d %H%M%S')

Le = LabelEncoder()
df['BS'] = Le.fit_transform(df['BS'])

df.head()
```

```python
X = df.drop(columns=['Energy','Time'])
Y = df['Energy']
model = LinearRegression()
split = X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Appliquer la régression polynomiale (degré 2 par exemple)
poly = PolynomialFeatures(degree=2)

# Transformation des features d'entraînement et de test
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Entraîner le modèle sur les données transformées
model = LinearRegression()
model.fit(X_train_poly, Y_train)

# Faire des prédictions
Y_pred = model.predict(X_test_poly)

# Évaluer la performance du modèle
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

# Afficher les résultats
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
