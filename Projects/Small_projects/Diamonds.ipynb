```python
import pandas as pd
import plotly.express as px
data = pd.read_csv('diamonds.csv')
data.head()
data.info()
data.drop(columns= 'Unnamed: 0', inplace=True)
#Group by carat and calculating the average price
avg_price_by_carat = data.groupby('carat')['price'].mean().reset_index()
#line chart
fig = px.line(avg_price_by_carat, x='carat', y='price', title='Average price of diamonds per carat', labels={'carat': 'Carat Size', 'price': 'Average Price'},
      template='plotly_white')
fig.show()
```
