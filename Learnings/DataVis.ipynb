```python
df = pd.read_csv("/content/used cars.csv")
```

Create a bar chart using Plotly :


```python
df
```

```python
fig = px.bar(df, x="Brand", y="Cars Listings", color="Cars Listings", height=400)
# color attribute colors the bars based on the "Cars Listings" values.
# The color scale will vary depending on the values, making it visually easier to distinguish higher/lower values.
# height=400, This sets the height of the figure to 400 pixels.
```

```python
fig
```

```python
fig.write_image("bar_chart.png")
```

Add a title and labels to the x and y axis


```python
fig.update_layout(title_text="Cars Listings by Brand", xaxis_title="Brand", yaxis_title="Cars Listings")
```

Customize the color scale and add a color bar :


```python
fig.update_traces(marker=dict(
    color='red',         # Sets the bar color to red
    opacity=0.7,         # Sets the transparency level (1 = fully opaque, 0 = fully transparent)
    line=dict(
        color='black',     # Sets the border (outline) color of the bars
        width=3        # Sets the thickness of the border (1.5 pixels)
    )
))
```

```python
fig.update_traces(marker=dict(color='red', opacity=0.7, line=dict(color='red', width=1.5)))
```

```python
fig = px.bar(df, x="Brand", y="Cars Listings", color="Cars Listings", height=400)
```

```python
fig.update_traces(marker=dict(
    color=df["Cars Listings"],  # Use actual data values
    opacity=0.7,
    line=dict(color='black', width=2.5)
))
```

```python
# Show the bar chart
fig.show()
```

## How to create pie chart using plotly :


```python
import plotly.express as px
import pandas as pd
# Define the data as a dictionary

data = pd.read_csv("/content/used cars.csv")

df = pd.DataFrame(data)
```

```python
df
```

```python
# Create a pie chart using Plotly Express, specifying the data frame, the column for the values, the column for the names,
#  and the color for each slice.

fig = px.pie(df, values='Cars Listings', names='Brand',
             color='Cars Listings', title='Cars')
```

```python
fig.show()
```

```python
# Update the text position of the labels to be inside each slice and show the percent and label information
# fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_traces(
    textposition='inside',   # Places text labels inside the bars or pie chart segments
    textinfo='percent+label' # Displays both the percentage value and the category label
)

fig.show()
```

```python
# Update the layout to set a minimum size for the text and to hide any overlapping text
# fig.update_layout(uniformtext_minsize=14, uniformtext_mode='hide')
fig.update_layout(
    uniformtext_minsize=14,   # Ensures all text labels have a minimum font size of 14
    uniformtext_mode='hide'   # Hides labels that don't fit within their respective sections
)

fig.show()
```
