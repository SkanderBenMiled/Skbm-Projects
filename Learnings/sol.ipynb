Load the dataset into a data frame using Python

```python
import pandas as pd
url = '/content/Africa_climate_change.csv'
df = pd.read_csv(url)
df.head()
```

Clean the data as needed.


```python
# Check for missing values
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# Convert the date column to datetime format if it's not already
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df.head(10
        )
```

Plot a line chart to show the average temperature fluctuations in Tunisia and Cameroon. Interpret the results.


```python
import plotly.express as px

# Filter data for Tunisia and Cameroon
df_filtered = df[df['COUNTRY'].isin(['Tunisia', 'Cameroon'])]

# Plot a line chart
fig = px.line(df_filtered, x='DATE', y='TAVG', color='COUNTRY', title='Average Temperature Fluctuations in Tunisia and Cameroon')
fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Average Temperature (°F)"
)
fig.show()

# 40 °F ~ 4 °C
# 72 °F ~ 22 °C
# 100 °F ~ 38 °C
```

Zoom in to only include data between 1980 and 2005, try to customize the axes labels.


```python
# Filter data between 1980 and 2005
df_zoomed = df_filtered[(df_filtered['DATE'].dt.year >= 1980) & (df_filtered['DATE'].dt.year <= 2005)]

# Plot the line chart with zoomed data
fig = px.line(df_zoomed, x='DATE', y='TAVG', color='COUNTRY', title='Average Temperature Fluctuations in Tunisia and Cameroon (1980-2005)')
fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Average Temperature (°F)"
)
fig.show()
```

Create Histograms to show temperature distribution in Senegal between [1980,2000] and [2000,2023] (in the same figure). Describe the obtained results.

```python
# Filter Senegal data
df_senegal = df[df['COUNTRY'] == 'Senegal']

# Data for the two periods
df_senegal_1980_2000 = df_senegal[(df_senegal['DATE'].dt.year >= 1980) & (df_senegal['DATE'].dt.year <= 2000)]
df_senegal_2000_2023 = df_senegal[(df_senegal['DATE'].dt.year >= 2000) & (df_senegal['DATE'].dt.year <= 2023)]

# Plot histograms
import plotly.graph_objects as go

fig = go.Figure()

# Histogram for 1980-2000
fig.add_trace(go.Histogram(x=df_senegal_1980_2000['TAVG'], name='1980-2000', opacity=0.75))

# Histogram for 2000-2023
fig.add_trace(go.Histogram(x=df_senegal_2000_2023['TAVG'], name='2000-2023', opacity=0.75))

# Update layout
fig.update_layout(
    barmode='overlay',
    title='Temperature Distribution in Senegal (1980-2000 vs 2000-2023)',
    xaxis_title="Average Temperature (°F)",
    yaxis_title="Frequency"
)
fig.show()
```

Select the best chart to show the Average temperature per country.


```python
# Group by country and calculate the average temperature
avg_temp_per_country = df.groupby('COUNTRY')['TAVG'].mean().reset_index()

# Create scatter plot
fig = px.scatter(avg_temp_per_country,
                 x='COUNTRY',
                 y='TAVG',
                 title='Average Temperature Per Country',
                 size_max=10)  # Optional: Adjust marker size

# Update layout for better readability
fig.update_layout(
    xaxis_title="Country",
    yaxis_title="Average Temperature (°F)",
    xaxis={'categoryorder': 'total descending'}  # Sort by descending average temp
)

fig.show()
```
