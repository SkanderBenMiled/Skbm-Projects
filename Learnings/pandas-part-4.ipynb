A z-score measures how many standard deviations a data point is from the mean.

Formula:
z = (x - μ) / σ

x: The data point.

μ: The mean of the column.

σ: The standard deviation of the column.

Let’s try to apply One hot encoding to the categorical feature “Role“ :



```python
data = pd.read_csv("/content/New_data (1).csv")
data.head(5)
```
