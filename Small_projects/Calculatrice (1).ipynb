```python
# on definie les fonction de calcul
def add(a, b):
  return a + b

def subtract(a, b):
  return a - b

def multiply(a , b):
  return a * b

def divide(a, b):
  if b == 0:
    return "Error: Division by 0 is not defined"
  else:
    return a / b
# premier message
def calculator():
  print("\n Lets calculate some number shall we ?")

  #dictionnaire

  operations = {"+": add, "-": subtract, "*": multiply, "/": divide}

  # premier numero

  num1 = float(input("Enter the first number :) : "))

  # condition
  for i in operations:
    print(i)
  should_continue = True

  while should_continue:
    operation_symbol = input("\nPick an operation:")
    if operation_symbol not in operations:
      print("Invalid operation. Please try again")
      continue
    num2 = float(input("Enter your second number :) "))
    # Operations
    calculation_fonction = operations[operation_symbol]
    result = calculation_fonction(num1, num2)
    print(f"{num1} {operation_symbol} {num2} = {result}")

    next = input("\n Type 'y' to continue with the result, 'n' to start a new 3amaleya :), q to quit").lower() # Si jamais une majuscule

    if next  == 'y': #continuer
      num1 = result
    elif next == 'n':
      calculator() # reboucle
      should_continue = False
    elif next == 'q':
      print("Barra 3ala khir") # Sortie de la boucle
      should_continue = False
    else:
      print("Invalid input. Please try again")
      should_continue = False

calculator()
```
